```python
import pandas as pd
import numpy as np

from dsdb import DsDbConnect
```

```python
df = pd.DataFrame(np.random.randint(1, 255, size=(200, 4)), columns=['a', 'b', 'c', 'd'])
df
```

```python
with DsDbConnect() as con:
    df.to_sql('test', con, index=False, if_exists='replace')
```

```python
with DsDbConnect() as con:
    df2 = pd.read_sql_table('test', con)
df2
```
